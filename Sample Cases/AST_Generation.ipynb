Explanation:
Iterative Implementation:

    left and right are the boundaries of the search.
    The while loop continues as long as left is less than or equal to right.
    mid is calculated as the middle index of the current search interval.
    If the target value is found at mid, the function returns the index.
    If the target value is greater than arr[mid], the search focuses on the right half (left = mid + 1).
    If the target value is smaller, the search focuses on the left half (right = mid - 1).
    If the target is not found, -1 is returned.
Recursive Implementation:

    The function calls itself with a smaller search interval based on the comparison with arr[mid].
    The base case checks if the search interval is invalid (left > right), meaning the target is not in the array.
    The process continues recursively until the target is found or the search interval becomes invalid.
Both implementations assume the input list is sorted.

```python
# Visiting Nodes from LinterCases.py - AddNumbers
class ASTVisitor(ast.NodeVisitor): #Extend the existing ast.NodeVisitor class
    # Extend with visit_<classname> of the body of the AST
    # For example if wanted to visit the Function Definition of a tree with a function declaration simply...
    # ...define visit_FunctionDef
    # Note: must dump the tree if planning on formatting as string
    # Method also requires self visits to recursively visit the node with the given classname
    def visit_FunctionDef(self, node):
        if node.__class__.__name__ == "FunctionDef":
            print(f"Function definition: {ast.dump(node)}")

        # print(node)
        self.generic_visit(node)  # Continue visiting child nodes
    
    def visit_Call(self, node):
        print(f"Function call: {ast.dump(node)}")
        self.generic_visit(node)  # Continue visiting child nodes
    
    def visit_Expr(self, node):
        print(f"Expression: {ast.dump(node)}")
        self.generic_visit(node)  # Continue visiting child nodes


# Instaniate the Node visitor
visitor = ASTVisitor()

# Run a ast tree through the Node Visitor
visitor.visit_FunctionDef(ast_tree)
```
